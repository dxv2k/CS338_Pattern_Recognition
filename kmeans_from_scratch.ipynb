# Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

# KMeans Class

```python
class KMeans():
    def __init__(self, n_clusters=5, tol = 0.0001):
        ''' 
        param: 
            n_clusters: number of K/Clusters using in Kmean
            tol: tolerance threshold, use to check consecutives iterations 
                                    to declare convergence 
        return: 
            None
        '''
        self.K = n_clusters
        self.tol = tol
        self.centroids = None

    def find_euclid_distance(self, a, b):
        return np.sqrt(np.sum(np.power((a - b), 2)))

    def _init_centroids(self): 
        # centroids random selection 
        rand_num = np.random.randint(0, len(X), size=self.K)
        self.centroids = np.array([X[i] for i in rand_num])
        
    def fit(self, X): 
        self._init_centroids()

        cluster = {}
        convergence = False
        while not convergence:
            # init empty cluster
            for i in np.arange(self.K):
                cluster[i] = []
            # iterate through data point in dataset
            for x in X:
                clusterth = 0
                min_dist = 0
                for j in range(self.K):
                    dist = self.find_euclid_distance(x, self.centroids[j])
                    if j == 0:
                        min_dist = dist
                        clusterth = 0
                    elif dist < min_dist:
                        min_dist = dist
                        clusterth = j
                cluster[clusterth].append(x)
            prev_centroids = self.centroids.copy()
            convergence = False
            # assign clusters & compute mean 
            for i in cluster:
                self.centroids[i] = np.average(cluster[i], axis=0)
                if (self.centroids[i] == prev_centroids[i]).all() == False:
                    sub = self.centroids[i] - prev_centroids[i]
                    # check for convergence
                    if all(j < self.tol for j in sub) == False:
                        convergence = True 

    def _predict_one(self, x):
        min_dist = 0
        clusterth = 0
        for j in range(len(self.centroids)):
            dist = self.find_euclid_distance(x, self.centroids[j])
            if j == 0:
                min_dist = dist
                clusterth = 0
            elif dist < min_dist:
                min_dist = dist
                clusterth = j
        return clusterth
    
    def predict(self, X): 
        return [self._predict_one(x) for x in X]
        
    def score(self, X, y):
        return sum(1 for predict, test in zip(self.predict(X), y) if predict == test) / len(y)
```

# Running the algorithm

```python
X, y = make_blobs(n_samples=1000, n_features=10, centers=5)
plt.rcParams.update({'figure.figsize':(10,7.5), 'figure.dpi':1000})
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=42)
```

```python
model = KMeans(n_clusters=5)
model.fit(X_train)
```

# Acccuracy of Kmeans 

```python
print('Result on train set: ',model.score(X_train,y_train))
print('Result on test set: ',model.score(X_test,y_test))
```
